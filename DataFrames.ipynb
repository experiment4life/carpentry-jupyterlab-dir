```python
# here we are using "numerical position" for both "row" and "column" using "iloc"
import pandas as pd
data = pd.read_csv("data/gapminder_gdp_europe.csv", index_col="country")
print(data.iloc[0,0]) # "position"[0,0] is first row (0), first column (0)
```

```python
# this is what the data looks like. here we can see Albania is first row (row zero)
import pandas as pd
data = pd.read_csv("data/gapminder_gdp_europe.csv", index_col="country")
print(data)
```

```python
print(data.loc["Albania", "gdpPercap_1952"])
```

```python
# to print only the one "row" for "Albania", 
# but print it for every column label 1952 -> 2007
# Similar syntax to "slicing a word string".

print(data.loc["Albania", :])
```

```python
# print out all the countries, but only the data under the column label "gdpPercap_1952"
print(data.loc[:,"gdpPercap_1952"])
```

```python
# we could also "ignore country name" and just specify the "column name" and it will
# produce all data under that "column"name.
print (data["gdpPercap_1952"])
```

```python
# show all "country rows" ...
# uses the colon ":" as the "slice" selection. ("from here to there")
# in this case, its all the rows [:]  
# but... only selecting from the column name "gdpPercap_1952"

print(data.loc[:, "gdpPercap_1952"])
```

```python
# to print multiple "columns" or multiple "rows" 
# you refer to these by their "label" and put a colon inbetween.
# from here "Italy" to there "Poland" (five countries)
# Note that the selection is "inclusive" at both ends 
# (unlike regular "slice" which is upto but not including)

print(data.loc["Italy":"Poland","gdpPercap_1962":"gdpPercap_1972"])

# as opposed to just one selection
# print(data.loc["Italy","gdpPercap_1962"])
```

```python
# We can chain commands together (like "|" in Bash), here we find the" max" value 
# we use the dot "." 
# the selection is the same , 
# but we add ".max()" at the end,note the call "()"
# and then we close off the print with a ")"

print(data.loc["Italy":"Poland", "gdpPercap_1962":"gdpPercap_1972"].max())
```

```python
# Using comparisons to select data based on value

# first get a small set to work with
subset = data.loc["Italy":"Poland", "gdpPercap_1962" : "gdpPercap_1972"]
print("Subset of data:\n" , subset)

# Which values were greater than 10000 ?
print("\nWhere are values large?\n", subset > 10000)
```
